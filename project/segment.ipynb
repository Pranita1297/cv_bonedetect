```python
from ultralytics import YOLO
import cv2
```

```python
# !pip install "ultralytics<=8.3.40" supervision roboflow
```

```python
model = YOLO("yolo11n.pt")

# train
model.train(
    data="C:\\Files for any work\\python university\\year2-2\\CV\\project\\bonev11\\data.yaml",
    epochs=100,
    imgsz=640,
    batch=8,
    name="bone-detection"
)
```

```python
model = YOLO("C:\\Files for any work\\python university\\year2-2\\CV\\project\\runs\\detect\\bone-detection3\\weights\\best.pt")
model("test.png", save=True, conf=0.3)

img = cv2.imread("test.png")
results = model(img, conf=0.3)

output = results[0].plot()
cv2.namedWindow("Result", cv2.WINDOW_NORMAL)
cv2.imshow("Result", output)

cv2.imshow("Result", results[0].plot())
cv2.waitKey(0)
cv2.destroyAllWindows()
```
